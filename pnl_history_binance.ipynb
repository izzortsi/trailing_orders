```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
thist = pd.read_excel("C:\\Users\\igor-\\Downloads\\transaction_history.xlsx")
pnl_hist = thist[thist["Type"] == "REALIZED_PNL"]
pnl_hist = pnl_hist.iloc[::-1]
pnl_hist.index = list(range(len(pnl_hist)))
pnl_hist = pnl_hist.rename(columns={"Date(UTC)": "Date"})
pnl_hist.Date = pnl_hist.Date.apply(pd.to_datetime)
```

```python
f, ax = plt.subplots(figsize=(16, 8))
ax.plot(pnl_hist.Date, pnl_hist.Amount, label="realized PNLs")
ax.plot(pnl_hist.Date, pnl_hist.Amount.cumsum(), label="cumulative realized PNLs")
ax.set_ylabel(r'return [\$]')
ax.legend()
```
